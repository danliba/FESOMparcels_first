```python
import uxarray as ux
import numpy as np
import xoak
from matplotlib import pyplot as plt
from cmocean import cm 
```

```python
path1 = "../../FESOM_data/channel/"
mesh_fn = "fesom.mesh.diag.nc"
u_path = 'u.fesom.2005.nc'
v_path = 'v.fesom.2005.nc'
#grid_path = "../../test/meshfiles/ugrid/outCSne30/outCSne30.ug"
#data_path = "../../test/meshfiles/ugrid/outCSne30/outCSne30_vortex.nc"
```

```python
uxds = ux.open_dataset(path1+mesh_fn, path1+u_path)
uxds
```

```python
uxds.uxgrid
```

```python
uxds.uxgrid.node_lon
```

```python
u=uxds["u"].compute()
u
```

```python
ux=u.isel(nz1=0).sel(time='2005-01-01')
ux
```

```python
ux=ux.isel(time=0, drop=True)
ux
```

```python
ux.plot(vmin=-1, vmax=1,
       cmap="coolwarm",
    figsize=(6, 10),
    add_colorbar=True,
    title="Zonal Velocity at the Surface (2005-01-01)")
```
